```python
# Loading important libraries
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
#%matplotlib inline sets the  backend of matplotlib to the inline backend
```

```python
# loading the dataset "iris"
iris = pd.read_excel(r'D:\CSRX BOX\iris.xlsx')
```

```python
# printing the first 5 rows
print(iris.head())
```

```python
# printing the last 5 rows
print(iris.tail())
```

```python
# We will go as
# heavy coding --> moderate coding --> no coding

# Scatter Plot using Matplotlib
fig,ax=plt.subplots()

# scatter the sepal length against sepal width
ax.scatter(iris['Petal_length'], iris['Sepal_width'])
# setting a title and labels
ax.set_title('Iris Datasets')
ax.set_xlabel('Sepal_length')
ax.set_ylabel('Sepal_width')
```

```python
# Line chart using Matplotlib
columns=iris.columns.drop(['Species_name'])
# creating x data
x_data=range(0, iris.shape[0])
# creating figure and axis
fig,ax=plt.subplots()
# plotting each column
for column in columns:
    ax.plot(x_data, iris[column], label=column)
# setting title and legend
ax.set_title('Iris Dataset')
ax.legend()
```

```python
# Scatter plot using pandas
iris.plot.scatter(x='Sepal_length', y='Sepal_width', title='Iris Dataset')
```

```python
# Scatter plot using seaborn without legend
import seaborn as sns
sns.scatterplot(x='Sepal_length', y='Sepal_width', data=iris)
```

```python
# Scatter plot using seaborn with legend
sns.scatterplot(x='Sepal_length', y='Sepal_width', hue='Species_name', data=iris)
```

```python
# Line chart using seaborn
sns.lineplot(data=iris.drop(['Species_name'], axis=1))
```

```python
# Finding the Kernel Density Plot for each type of species(FacetGrid) using seaborn
g = sns.FacetGrid(iris, col='Species_name')
g = g.map(sns.kdeplot, 'Sepal_length')
```

```python
# Drpping column "Species_name" so that remaining columns be numeric to get correlation matrix
iris=iris.drop(['Species_name'],axis=1)
```

```python
# Getting correlation matrix
corr = iris.corr()
fig, ax = plt.subplots()
# creating heatmap
im = ax.imshow(corr.values)

# setting labels
ax.set_xticks(np.arange(len(corr.columns)))
ax.set_yticks(np.arange(len(corr.columns)))
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        text = ax.text(j, i, np.around(corr.iloc[i, j], decimals=2),ha="center", va="center", color="black")
```

```python
# Heatmap using seaborn
sns.heatmap(iris.corr(), annot = True)
```

```python
# Pairplot using seaborn
sns.pairplot(iris)
```
