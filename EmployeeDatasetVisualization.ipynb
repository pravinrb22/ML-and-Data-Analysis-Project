```python
# Loading important libraries
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import sweetviz as sv
import dtale
%matplotlib inline
#%matplotlib inline sets the  backend of matplotlib to the inline backend
```

```python
# loading the dataset "Employee data"
employee_data = pd.read_csv(r'D:\Pravin_folder\CSRX BOX\Employee data.csv')
```

```python
# Printing the first 5 rows of the data set
print(employee_data.head())
```

```python
# Printing the last 5 rows of the data set
print(employee_data.tail())
```

```python
# To find the summary statistics
print(employee_data.describe())
```

```python
# plotting Bar chart for employee distribution by job categories

#creating a the figure with a size of 10 inches by 6 inches
plt.figure(figsize=(10, 6))

# plotting the chart using seaborn
sns.countplot(x='jobcat', data=employee_data)

# Adding levels and titles
plt.title('Employee Distribution by Job Categories')
plt.xlabel('Job Categories')
plt.ylabel('Number of Employees')

# To display the chart
plt.show()
```

**Conclusion:** The bar chart reveals that the majority of employees are concentrated in specific jo categories.

```python
# Count plot for Gender Distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='gender', data=employee_data, palette={'m': 'green', 'f': 'blue'})
plt.legend(labels=['Male', 'Female'], title='Gender')
plt.title('Gender Distribution of Employees')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

**Conclusion:** Number of Male workers are higher than the number of Female workers.

```python
# Calculating the average salary for each gender
AvgSalaryByGender = employee_data.groupby('gender')['salary'].mean()

# Plotting the bar plot
plt.figure(figsize=(8, 5))
AvgSalaryByGender.plot(kind='bar', color=['pink','blue'], edgecolor='black')
# Adding labels and titles
plt.title('Average Salary by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Salary')

#  To Display the bar plot
plt.show()
```

**Conclusion:** Average salary of male workers is higher than the average salary of female workers.

```python
# Pie chart for gender distribution
gender_distribution = employee_data['gender'].value_counts()
plt.figure(figsize=(6, 6))
labels = ['Male', 'Female']
plt.pie(gender_distribution, labels=gender_distribution.index, autopct='%1.1f%%', startangle=90)
plt.legend(labels, title='Gender', loc='upper right')
plt.title('Gender Distribution of Employees')
plt.show()
```

```python
# Pie chart to visualize the distribution of educational backgrounds among employees

# Count the occurrences of each education level
education_counts = employee_data['educ'].value_counts()

# Plotting the pie chart with a different color palette and improved formatting
plt.figure(figsize=(8, 8))

# Choose a different color palette
colors = plt.cm.Set3.colors 
plt.pie(education_counts, labels=education_counts.index,autopct='%1.1f%%', startangle=90, colors=colors,wedgeprops=dict(width=0.7),labeldistance=1.15,pctdistance=0.85)

# Adding title
plt.title('Distribution of Educational Backgrounds of Employees')

# Display the pie chart
plt.show()
```

```python
# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(employee_data['salary'], bins=20, color='pink',edgecolor='black')

# Adding labels and title
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')

# Display the histogram
plt.show()
```

```python
# Histogram for Employees Education
plt.figure(figsize=(10, 6))
sns.histplot(employee_data['educ'], bins=20, kde=True, color='blue')
plt.title('Education of Employees')
plt.xlabel('Education')
plt.ylabel('Frequency')
plt.show()
```

```python
# scatter plot for salary distribution by work experience
plt.figure(figsize=(11, 6))
sns.scatterplot(x='prevexp', y='salary', hue = 'gender', data=employee_data, s = 50)
plt.title('Salary Distribution by Previous Experience')
plt.xlabel('Previous Work Experience(years)')
plt.ylabel('Salary')
plt.show()
```

```python
# Box plot for salary distribution by Education Level
plt.figure(figsize=(12, 8))
sns.boxplot(x='educ', y='salary', data=employee_data)
plt.title('Salary Distribution by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Salary')
plt.show()
```

```python
# Boxplot for Salary Distribution of  Employees
plt.figure(figsize=(8, 6))
sns.boxplot(x='salary', data=employee_data)
plt.title('Salary Distribution of Employees')
plt.xlabel('Salary')
plt.show()
```

```python
# Boxplot of Salary Distribution by Job Categories
plt.figure(figsize=(10, 6))
sns.boxplot(x='jobcat', y='salary', data=employee_data)
plt.title('Salary Distribution by Job Categories')
plt.xlabel('Job Categories')
plt.ylabel('Salary')
plt.show()
```

**Conclusion:** The box plot illustrates the variation in salary distribution across different jo categories, helping identify potential disparities and informing HR policies related to compensation.

```python
# Line chart for employee count over years of experience
experience_count = employee_data.groupby('prevexp')['id'].count().reset_index()
plt.figure(figsize=(10, 5))
plt.plot(experience_count['prevexp'], experience_count['id'], marker='o')
plt.title('Employee Count by Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Number of Employees')
plt.grid(True)
plt.show()
```

**Conclusion:** The line chart demonstrates the distribution of employees based on their years of experience, offering insights into the organization's workforce composition and potential trends related to employee retention or turnover.

```python
# Pairplot for relationship between Numerical Variables
sns.pairplot(employee_data[['educ', 'salary', 'prevexp']], height= 3)
plt.suptitle('Pairplot of Numerical Variables', y=1.02)
plt.show()
```

```python
# Dropping columns "gender" and "bdate"
df = employee_data.drop(['gender', 'bdate'], axis = 1)
```

```python
# Getting correlation matrix
corr = df.corr()
fig, ax = plt.subplots()
# creating heatmap
im = ax.imshow(corr.values)

# setting labels
ax.set_xticks(np.arange(len(corr.columns)))
ax.set_yticks(np.arange(len(corr.columns)))
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        text = ax.text(j, i, np.around(corr.iloc[i, j], decimals=2),ha="center", va="center", color="black")
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```
